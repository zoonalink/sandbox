# Data Modelling Development Plan

## 1. CDMP (Certified Data Management Professional)

**Priority:** HIGH

**Focus Areas:**
- Data Modeling & Design
- Data Architecture
- Data Integration & Interoperability
- Data Governance

**Approach:**
- Exams: https://cdmp.info/exams/
- https://cdmp.info/about/
- CDMP Fundamentals exam (requires 60% to pass; 70% for practitioner) - $311 
- Progress to Practitioner level with specialisation in two out of the following areas (each $311):
  - Data Modeling and Design 
  - Metadata
  - Data Quality
  - Data Governance
  - Data Warehousing & Business Intelligence
  - Reference & Master Data Management
  - Data Integration & Interoperability
- The certification consists of two exams:
  * Core exam covering all DMBOK knowledge areas
  * Specialty exam focusing on Data Modeling

**Resources:**
- Official Website: https://cdmp.info/
- Study Guide: DAMA-DMBOK (Data Management Body of Knowledge), latest edition
  - https://technicspub.com/dmbok2/
- Exam cost: $311-$695 depending on level and DAMA membership status
- Practice exams available through DAMA International

**Requirements:**
- Minimum 2-3 months study time (10-15 hours/week)
- DAMA membership recommended ($60-$85 annually)
- Preparation courses available but not required
- Certification valid for 3 years before renewal

## 2. Data Modelling Tool Certifications

**Priority:** HIGH

**Options:**

1. **ERwin Data Modeler Certification**
   - Official site: https://erwin.com/education/
   - Focused on enterprise data modeling capabilities
   - Widely used in higher education institutions
   - Levels: Associate, Professional, Master
   - Cost: $200-$250 per exam
   - **HEI recommendation:** Strong choice for large universities with complex data landscapes

2. **ER/Studio Data Architect Professional (EDAP)**
   - Official site: https://www.idera.com/services/education/er-studio
   - Comprehensive tool for enterprise data modeling
   - Good integration with multiple database platforms
   - Cost: $150-$200 per exam
   - **HEI recommendation:** Good alternative if cost is a primary concern

3. **SAP PowerDesigner Certification**
   - Official site: https://www.sap.com/products/powerdesigner-data-modeling-tools.html
   - Enterprise-class modeling and metadata management
   - Strong in business process modeling integration
   - Cost: $500-$700
   - **HEI recommendation:** Best for institutions with existing SAP infrastructure

**HEI Enterprise Tool Recommendation:**
ERwin Data Modeler is particularly well-suited for higher education institutions due to its:
- Strong support for both logical and physical data models
- Integration capabilities with BI tools commonly used in universities
- Collaborative features for stakeholder involvement
- Metadata management capabilities important for regulatory compliance

## 3. Dimensional Modelling (Kimball Methodology)

**Priority:** MEDIUM-HIGH

**Focus Areas:**
- Star schema design
- Snowflake schema variations
- Slowly changing dimensions (SCDs)
- ETL process design
- Data warehouse architecture

**Resources:**
- Book: "The Data Warehouse Toolkit" by Ralph Kimball and Margy Ross
  https://www.kimballgroup.com/data-warehouse-business-intelligence-resources/books/data-warehouse-dw-toolkit/
- Kimball Group courses: https://www.kimballgroup.com/data-warehouse-business-intelligence-resources/
- eLearningCurve Data Warehousing courses: https://elearningcurve.com/
- Dimensional modeling certification through The Data Warehouse Institute (TDWI)
  https://tdwi.org/Home.aspx

**Implementation:**
- No single official certification, but recognized through:
  * TDWI Data Warehousing Concepts certification
  * Kimball University training (when available)
  * University of Washington certificate program in data warehousing

## 4. TOGAF Data Architecture

**Priority:** MEDIUM

**Value:**
- Connects data models to broader enterprise architecture
- Provides framework for aligning data architecture with university strategic objectives
- Helps communicate data modeling decisions to non-technical stakeholders
- Supports data governance implementation

**Approach:**
- TOGAF 9.2 Foundation certification (Level 1)
- TOGAF 9.2 Certified (Level 2)
- Focus on data architecture components within TOGAF framework

**Links:**
- Official site: https://www.opengroup.org/togaf
- Certification details: https://www.opengroup.org/certifications/togaf
- Exam registration: https://www.theopengroup.org/certifications/how-to-get-certified
- Study guides: https://publications.opengroup.org/certification-guides

**Requirements:**
- Foundation level: 40 question exam, 55% passing score
- Certified level: Scenario-based exam
- Combined cost: $550-$750
- Training courses available from accredited providers

## 5. Others - Data Vault, Cloud Data, etc.

### Data Vault 2.0 Certification

**Priority:** MEDIUM

**Details:**
- Modern approach to enterprise data warehousing
- Focuses on scalability, adaptability, and consistency
- Official site: https://datavaultalliance.com/certification/
- Training options: https://www.learndatavault.com/
- Cost: $1,200-$2,000 for bootcamp and certification
- Particularly valuable for universities managing research data and student information across multiple systems

### Cloud Data Architecture Certifications

**Priority:** MEDIUM

**Options:**

1. **AWS Database Specialty**
   - https://aws.amazon.com/certification/certified-database-specialty/
   - Cost: $300
   - Focus: Amazon RDS, DynamoDB, Redshift, etc.

2. **Microsoft Certified: Azure Data Engineer Associate**
   - https://learn.microsoft.com/en-us/certifications/azure-data-engineer/
   - Cost: $165
   - Focus: Azure SQL, Data Factory, Synapse Analytics, etc.

3. **Google Professional Data Engineer**
   - https://cloud.google.com/certification/data-engineer
   - Cost: $200
   - Focus: BigQuery, Dataflow, Dataproc, etc.

**HEI Recommendation:** 
- Choose based on university's cloud strategy
- Azure often preferred in UK higher education settings
